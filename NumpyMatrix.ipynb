```python
data = np.array([10,15,17,20,25,26,28,30,32,40,45,70,76,80,100])
```

```python
mean = np.mean(data)
mean # Ortalama
```

```python
std = np.std(data)
std # Standart Sapma
```

```python
z_scores = (data - mean) / std
z_scores
```
