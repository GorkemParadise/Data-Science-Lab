```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.Figure(figsize=(6, 5))
sns.boxplot(y=data)
plt.title('Box Plot Example')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("12-TitanicData.csv")
```

```python
df
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data=df)
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age") 

plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data=df)
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```
