```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10, 20, 25, 30, 35, 40, 45, 50, 60]
weight_list = [70, 65, 70, 75, 90, 85, 90, 100, 85]
```

```python
plt.plot(age_list, weight_list, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.savefig("test.png")
plt.show()
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list, np_weight_list, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,19)
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b*-")
plt.show()
```

```python
plt.plot(numpy_arr2, numpy_arr1, "y*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2, "y*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1, "g*-")
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3])
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show
```

```python
new_fig = plt.figure(dpi=200)
new_axes = new_fig.add_axes([0.1,0.1, 0.9,0.9])
new_axes.plot(numpy_arr1, numpy_arr1 ** 2, label = "arr ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3, label = "arr ** 3")
new_axes.legend()
plt.show()
```
