```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("11-Athlete_events.csv")
```

```python
data.describe()
```

```python
data.isnull()
```

```python
data.info()
```

```python
plt.scatter("Height", "Weight", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Boy vs Ağırlık")
```

```python
sns.set_theme(style="darkgrid")
sns.scatterplot(x="Height", y="Weight", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.lineplot(x="Height", y="Weight", hue="Sex", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.displot(x="Height", hue="Sex", data=data, kind="kde", fill=True)
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal", y="Height", hue="Sex", data=data)
plt.title("Medal Height")
plt.show()
```

```python
sns.catplot(x="Medal", y="Height", hue="Sex", col="Season", data=data)
```
